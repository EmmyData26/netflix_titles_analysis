## Data Visualization

- Visual representation of data to discover trends and patterns
- Helps in storytelling and decision-making

### Types

- Categorical: Bar Chart, Pie Chart, e.t.c
- Numerical: Box plot, Histogram, e.t.c
- Relationship: Scatter plot, Line plot, Heatmap, Pair plot, e.t.c

### Libraries

- Matplotlib
- Seaborn

```python
import pandas as pd

import matplotlib.pyplot as plt
```

- Figure: The entire canvas that holds all subplots (The umbrella covering all plots)
- Axes: A single plot

```python
#Create a Figure with a single axes

x = [1,2,3]

y = [4,5,6]

#We have to initialise the figure and axes

fig, ax=plt.subplots()
ax.plot(x,y)
ax.set_title("Basic Plot") #To give title to my plot
ax.set_xlabel("X-axis")
ax.set_ylabel("Y- Axis")

plt.show()#to display the plots
```

```python
#Create a figure with multiple subplots

fig, axes=plt.subplots(1,2) #Two plots side by side
axes[0].plot(x,y)
axes[0].set_title("Left Plot")
axes[1].plot(x,y)
axes[1].set_title("Right Plot")
plt.show()
```

```python
fig, axes = plt.subplots(2,2)

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(x,y)

plt.show()
```

What are the  number of netflix titles that have being released per year?

#We have to prepare the data before we plot, either by pivot, grouping

```python
title_per_year = df.groupby("release_year")["title"].count().reset_index()

title_per_year
```

```python
#since both x and y axis are numerical, the best is to use a line plot

plt.plot(title_per_year["release_year"], title_per_year["title"], color = "red", marker = "o")
plt.title("Number of NetFlix Title Released Per Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

Bivariate Analysis

- Examining the relationship etween two variables

```python
df["type"].value_counts()
```

```python
#Movies and TV shows released per year

release_year_by_type = df.groupby(["release_year", "type"]).size() #size helps to count it per each group

release_year_by_type
```

The way it is, is a series object

```python
release_year_by_type = df.groupby(["release_year", "type"]).size().unstack(fill_value=0).reset_index() 

release_year_by_type
```

```python
plt.plot(release_year_by_type["release_year"], release_year_by_type["Movie"], label = "Movies")
plt.plot(release_year_by_type["release_year"], release_year_by_type["TV Show"],label = "TV Shows")
plt.legend()


plt.show()
```
